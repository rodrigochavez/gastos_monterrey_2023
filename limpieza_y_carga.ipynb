```python
import pandas as pd
```

```python
df_gastos_varios = pd.read_excel('/content/drive/MyDrive/pagos_provedores_monterrey/DICIEMBRE_2023_VP.xlsx', sheet_name='GTOS VARIOS', skiprows=5)
df_gastos_varios['SHEETS'] = 'GTOS VARIOS'
```

```python
df_contratistas = pd.read_excel('/content/drive/MyDrive/pagos_provedores_monterrey/DICIEMBRE_2023_VP.xlsx', sheet_name='OBRA PUBLICA Y GV FDOS. ESP', skiprows=5)
df_contratistas['SHEETS'] = 'OBRA PUBLICA Y GV FDOS. ESP'
```

```python
df_servicios = pd.read_excel('/content/drive/MyDrive/pagos_provedores_monterrey/DICIEMBRE_2023_VP.xlsx', sheet_name='SER PROF', skiprows=5)
df_servicios['SHEETS'] = 'SER PROF'
```
